# Introdução

Desde os príncipios da humanidade, já se usavam constelações e estudos dos astros para localizações e calendários. Ao procurar entender o Universo, o ser humano desenvolveu diversas tecnologias e pesquisas que de alguma maneira passaram também a fazer parte do seu cotidiano, alguns exemplos são : GPS, painéis solares, satélites de comunicação.

O universo está sendo aos poucos explorado com ajuda de telescópios, sendo o mais famoso o telescópio espacial Hubble, lançado em 1990. A visão do Hubble, permitiu a humanidade pela primeira vez observar corpos celestes além da nossa galáxia.

Existem diversos corpos celestes que um telescópido consegue observar, entre eles os mais importantes são aqueles que exercem uma estrutura gravitacional em sua volta. Iremos abordar 3 principais: estrelas; quasar; galáxias.

### Estrelas
As estrelas são esferas de plasma mantidas por sua própria gravidade, podendo conter sistemas de outros corpos celestes sob seu domínio gravitacional. As estrelas mais importantes da esfer celeste da Terra, foram agrupadas em constelações e asterimos, podendo ser identificadas no céu noturno. Elas podem ser classificadas por temperatura e efeitos de luminosidade, por exemplo as estrelas anãs brancas possuem sua própria classificação diferente das outras estrelas.

### Quasares
O Quasar é um objeto astronômico que possui um centro gravitacional maior que as estrelas porém ele não tem o mínimo para ser considerado uma galáxia. Porém eles são os maiores emissores de energida do Universo, brilhando até mil vezes mais que uma galáxia inteira. Não existia consenso sobre esses objetos até os anos 80.


### Galáxias
A Galáxia é um sistema gravitacionalmente ligado de estrelas e portanto é a estrutura gravitacional mais poderosa que iremos abordar. O sistema solar faz parte da galáxia chamada de Via Láctea, e esta é observada desde a antiguidade. As galáxias são categorizadas por sua forma visual, sendo a mais comum em formato elíptico. Um tipo excepcional de gálaxias são as starbusters, que possuem um número elevado de formação de estrelas.

### Importância do estudo dos astros
É interessante para estudos astronômicos, analisar as informações extraídas através de telescópios por meio de sistemas fotométricos e identificar qual tipo de corpo é a estrutura gravitacional encontrada. O eletroscópio de um telescópio, por exemplo, consegue medir o desvio para o vermelho, isto é a alteração da frequência de luz emitida da fonte (astro) ao receptor (Terra). Este desvio pode ser causado pelo Efeito Doppler ou, então, devido aos campos gravitacionais pela qual a luz passa ou até mesmo, pela expansão do Universo (denominado redshift cosmológico).

Porém, por serem estruturas gravitacionais, apenas analisar o desvio para o vermelho de um astro não basta. Observe que uma luz emitida por uma estrela será afetada pelo campo gravitacional da estrela e pelo campo gravitacional da galáxia que ela pertence. 

Portanto, o objetivo deste trabalho é conseguir um classificador confiável que não apenas facilite o trabalho de identifição de corpos celestes nas três categorias abordadas mas, também, auxilie astrônomos a encontrar novos padrões, comportamentos e diversas outras informações úteis que possam gerar um conhecimento poderoso para a humanidade

# Descrição dos Atributos 

- **objid**: Corresponde ao identificador do objeto.


- **ra (Right Ascension)**: A Ascensão Reta é uma das coordenadas do sistema de coordenadas equatorial que corresponde a distância angular, medida no Equador Celeste, entre um ponto de origem e o astro. Durante muitos anos, o ponto de origem utilizado era a localização do centro do Sol durante o equinócio de Março. Contudo, atualmente a origem da ascensão reta é o quasar 3C 273, através do catálogo FK5. A Ascesão Reta, é análoga à coordenada geográfica Longitude.
    A ascensão reta pode ser medida tanto em horas quanto em graus, sendo o intervalo em graus igual a $[0, 360]$


- **dec (Declination)**: A Declinação é a segunda coordenada do sistema de coordenadas equatorial, correspondendo ao ângulo do equador celeste ao astro em questão. A Declinação é comumente medida em graus em um intervalo de -90 a 90 graus, onde o grau negativo corresponde as localizações sobre o Hemisfério Sul e graus positivos sobre o Hemisfério Norte


- **(u,g,r,i,z)**: A Magnitude é uma escala logarítmica utilizada para definir o brilho de um astro. Esse brilho pode ser captado através do comprimento de onda que chega no instrumento de medição ou, através da banda passante (filtro) correspondente. Por existirem padrões de frequência na incidência de cor e brilho, foram desenvolvidos sistemas fotométricos. Um sistema fotométrico pode ser definido como um conjunto de filtros bem definidos (ou seja, padrão de incidências em certas cores). 
    O Sistema Fotométrico utilizado pelos observadores do Sloan Digital Sky Survey é o Thuan-Gunn Astronomic Magnitude System, onde **u, g, r, i, z** correspondem as 5 bandas passantes (filtros) do telescópio.
    - u: banda ultravioleta
    - g: banda verde
    - r: banda vermelha
    - i: banda infravermelha
    - z: banda inferior ao infravermelho
    
    
- **(run, rerun, camcol, field)**: Atributos que correspondem as características dos instrumentos utilizados, como, telescópio, caâmera e espectrômetro.


- **class (Classe)**: Classe do objeto sendo este, uma estrela, galáxia ou quasar.


- **redshift (Desvio para vermelho)**: O desvio para vermelho corresponde a alteração na frequência da luz do emissor até o receptor. A causa da alteração pode ser o efeito Doppler, interação com outros campos gravitacionais (mais comum) e a expansão do Universo (redshift cosmológico)


- **(plate, mjd, fiberid)**: Atributos que correspondem as características dos instrumentos utilizados, no caso, um espectroscópico

Por fim, uma vez que ruídos foram removidos e os dados processados, realizamos a etapa de mineração e desenvolvimento do classificador. É interessante ressaltar que existem alguns atributos mais e menos relevantes. Logo, decidimos desenvolver diferentes modelos para diferentes conjuntos de atributos utilizados.

Isto é, foram selecionados alguns conjuntos de atributos e formado um conjunto de seleção de modelo (treinamento) e teste, sendo 70% do conjunto de dados total dedicado ao treinamento e 30% ao teste. Em seguida, ocorreu o processo de seleção do modelo, onde os melhores hiperparâmetros foram definidos através de um processo de 10-Fold Cross Validation.

Tal processo se deu, pois o ajuste de hiperparâmetros de um modelo de Aprendizado de Máquina sobre um conjunto de dados durante o treinamento, aumenta muito as chances de se produzir um overfitting. 

Após definidos os hiperparâmetros, os modelos produzidos foram submetidos ao processo de teste tendo como métricas de avaliação: Acurácia, F1-Score, Precisão e Revocação.

Os modelos adotados foram: k-Nearest Neighbor (kNN), Naive Bayes (Gaussian Naive Bayes) e Support Vector Machine (com kernel RBF).

Ou seja, o processo de mineração se deu da seguinte forma.

1. Selecionado atributos mais relevantes e definidos três conjuntos de dados
    1. Atributos **u, g, r, i, z**
    2. Atributos **ra, dec, u, g, r, i, z, redshift**
    3. Atributos **ra, dec, u, g, r, i, z, redshift** padronizado (distribuição com média 0 e desvio padrão 1)

2. Para cada um dos três conjunto de dados formado, foi feito o mesmo processo de treinamento e teste.
    1. Divisão do conjunto de dados em 70% de treinamento e 30% de teste
    2. O conjunto de treinamento foi utilizado para o processo de 10-Fold Cross Validation sendo, assim, dividido em 10 partições, onde 9 são tilizadas para treino e 1 utilizada para teste.
    3. Durante o 10-Cross Validation, os modelos tiveram seu hiperparâmetros ajustados
        - kNN: Número de vizinhos de 1 a 9.
        - SVM: C (Penalidade) de 1.0 a 9.0.
    4. Após encontrado o melhor valor para o hiperparâmetro, os modelos são treinados em todo o conjunto de treinamento que durante o 10-Fold foi dividido em 10 partições e testados no conjunto de teste correspondente a 30% do conjunto de dados total.

```python
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn.model_selection import StratifiedKFold
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score

np.random.seed(1)
```

```python
encoder = preprocessing.LabelEncoder()
Y = encoder.fit_transform(dt['class'].values)
X = dt.iloc[:,3:8].values

kNN_acc = []
kNN_f1s = []
kNN_pre = []
kNN_rec = []

nb_acc  = []
nb_f1s  = []
nb_pre  = []
nb_rec  = []

svm_acc = []
svm_f1s = []
svm_pre = []
svm_rec = []
```

# Conjunto de Dados 1

```python
x, x_test, y, y_test = train_test_split(X, Y, test_size=0.3)
kFold = StratifiedKFold(n_splits=10)
for train, test in kFold.split(x, y):
    kNN = None
    kNN = KNeighborsClassifier(n_neighbors=7)
    kNN.fit(x[train], y[train])
    nb = None
    nb = GaussianNB()
    nb.fit(x[train], y[train])
    svm = None
    svm = SVC(C=2.5, kernel='rbf', gamma='auto')
    svm.fit(x[train], y[train])
    
    y_pred_kNN = kNN.predict(x[test])

    kNN_acc.append(accuracy_score(y[test],  y_pred_kNN))
    kNN_f1s.append(f1_score(y[test],        y_pred_kNN, average='macro'))
    kNN_pre.append(precision_score(y[test], y_pred_kNN, average='macro'))
    kNN_rec.append(recall_score(y[test],    y_pred_kNN, average='macro'))

    y_pred_nb = nb.predict(x[test])

    nb_acc.append(accuracy_score(y[test],  y_pred_nb))
    nb_f1s.append(f1_score(y[test],        y_pred_nb, average='macro'))
    nb_pre.append(precision_score(y[test], y_pred_nb, average='macro'))
    nb_rec.append(recall_score(y[test],    y_pred_nb, average='macro'))

    y_pred_svm = svm.predict(x[test])
    
    svm_acc.append(accuracy_score(y[test],  y_pred_svm))
    svm_f1s.append(f1_score(y[test],        y_pred_svm, average='macro'))
    svm_pre.append(precision_score(y[test], y_pred_svm, average='macro'))
    svm_rec.append(recall_score(y[test],    y_pred_svm, average='macro'))
```

```python
print('7-Nearest neighbor')
print("acc: %.3f +/- %.3f\nf1s: %.3f +/- %.3f\n"
      "pre: %.3f +/- %.3f\nrec: %.3f +/- %.3f"  % 
     (np.mean(kNN_acc), np.std(kNN_acc), np.mean(kNN_f1s), np.std(kNN_f1s), 
      np.mean(kNN_pre), np.std(kNN_pre), np.mean(kNN_rec), np.std(kNN_rec)))
```

```python
print('Gaussian Naive Bayes')
print("acc: %.3f +/- %.3f\nf1s: %.3f +/- %.3f\n"
      "pre: %.3f +/- %.3f\nrec: %.3f +/- %.3f"  % 
     (np.mean(nb_acc), np.std(nb_acc), np.mean(nb_f1s), np.std(nb_f1s), 
      np.mean(nb_pre), np.std(nb_pre), np.mean(nb_rec), np.std(nb_rec)))
```

```python
print('Support Vector Machine (RBF Kernel)')
print("acc: %.3f +/- %.3f\nf1s: %.3f +/- %.3f\n"
      "pre: %.3f +/- %.3f\nrec: %.3f +/- %.3f"  % 
     (np.mean(svm_acc), np.std(svm_acc), np.mean(svm_f1s), np.std(svm_f1s), 
      np.mean(svm_pre), np.std(svm_pre), np.mean(svm_rec), np.std(svm_rec)))
```
